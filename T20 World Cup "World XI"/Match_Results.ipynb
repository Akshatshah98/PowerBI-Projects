```python
#Preparing Data for preprocessing and Exploratory Data Analysis
#Import libraries
import pandas as pd
import json
```

# Processing all the match results

```python
#Retrieving Data from the json files

with open('json_files/Match_results.json') as f:
    data = json.load(f)

df_match_results = pd.DataFrame(data[0]['matchSummary'])
df_match_results.head(20)
```

```python
#Description of the followinf dataset
df_match_results.describe()
```

```python
df_match_results.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match_results.head()
```

```python
id_dict = {}

for index, row in df_match_results.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    id_dict[key1] = row['match_id']
    id_dict[key2] = row['match_id']
```

```python
df_match_results.to_csv('csv_files/Match_Reults.csv', index = False)
```

# Processing all the Batting Data

```python
with open('json_files/Batting_info.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting_info = pd.DataFrame(all_records)
df_batting_info.head(20)
```

```python
df_batting_info['out/not_out'] = df_batting_info.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting_info.head(20)
```

```python
df_batting_info['match_id'] = df_batting_info['match'].map(id_dict)
df_batting_info.head(20)
```

```python
df_batting_info.drop(columns=["dismissal"], inplace=True)
df_batting_info.head(20)
```

```python
#Now, the problem with this dataset is there are few unrecognized or unused in the data, which are to be removed.
#Hence, we'll do some data cleaning.
```

# Cleaning the dataset

```python
df_batting_info['batsmanName'] = df_batting_info['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting_info['batsmanName'] = df_batting_info['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting_info.head(20)
```

```python
#Exporting the batting data to a csv file.
df_batting_info.to_csv('csv_files/Bating_Data.csv', index = False)
```

# Processing all the Bowling Data

```python
with open('json_files/Bowling_Info.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
df_bowling_info = pd.DataFrame(all_records)
df_bowling_info.head(20)
```

```python
df_bowling_info['match_id'] = df_bowling_info['match'].map(id_dict)
df_bowling_info.head(20)
```

```python
#The Bowling data seems pretty much clean. And there is no such descripency to look out for. Hence we'll just export
#this data to a csv file.

df_bowling_info.to_csv('csv_files/Bowling_Data.csv', index = False)
```

# Processing all the Individual Player's Information

```python
with open('json_files/Player_info.json') as f:
    data = json.load(f)
```

```python
df_players_info = pd.DataFrame(data)
df_players_info.head(20)
```

```python
#As it seems there are some descripencies to resolve. Let's clean the data further.
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('â€', ''))
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('†', ''))
df_players_info['name'] = df_players_info['name'].apply(lambda x: x.replace('\xa0', ''))
df_players_info.head(20)

#Let's just gather the information for the players from Team India.
df_players_info[df_players_info['team'] == 'India']
```

```python
#Exporting the Indian players' information to a csv file.

df_players_info.to_csv('csv_files/Players_Data.csv', index = False)
```
